## Remember...
### When we are given 2 points on a graph, the formula below is used to calculate and graph the slope of its line:

$$slope = {y2-y1 \over x2-x1}$$

#### NOTES* A slope of 0 with give you a vertical line.
#### NOTES* A 0 in the denominator of your slope (run) will give a line called UNDEFINED

#### Take a look at number 5.  Our coordinates are (1, 2), and (7,7).  Before we continue, lable X1, Y1, X2, and Y2 on your paper.  It should look like this...
###  X1= 1
### Y1= 2
### X2= 7
### Y2= 7

Plug in each number into our slope formula and use your calculator to solve: $$slope = {y2-y1 \over x2-x1}$$

### slope = 7 - 2 \ 7 - 1

### Answer: 5/6

Did you get the right answer?  Now lets check the same problem using a Python code...

```python
import matplotlib.pyplot as plt

def calculate_slope(x1, y1, x2, y2):
    delta_y = y2 - y1
    delta_x = x2 - x1
    slope = delta_y / delta_x
    return slope

# Prompt the user to input the coordinates of the two points
x1 = float(input("Enter the x-coordinate of the first point: "))
y1 = float(input("Enter the y-coordinate of the first point: "))
x2 = float(input("Enter the x-coordinate of the second point: "))
y2 = float(input("Enter the y-coordinate of the second point: "))

# Call the function to calculate the slope
slope = calculate_slope(x1, y1, x2, y2)

if slope == float('inf'):
    print("The line is vertical. The slope is undefined.")
else:
    print("The slope of the line passing through the two points is:", slope)

# Generate x-values for the line
x_values = [x1, x2]

# Generate y-values for the line using the slope-intercept form of the line equation: y = mx + c
y_values = [slope * x + y1 - slope * x1 for x in x_values]

# Create a figure and axis
fig, ax = plt.subplots()

# Plot the line
ax.plot(x_values, y_values, label='Line', color='b')

# Plot the two points
ax.scatter([x1, x2], [y1, y2], label='Points', color='r')

# Add annotations for the coordinates
ax.annotate(f"({x1}, {y1})", (x1, y1), textcoords="offset points", xytext=(-15, 10), ha='center', fontsize=8)
ax.annotate(f"({x2}, {y2})", (x2, y2), textcoords="offset points", xytext=(10, -15), ha='center', fontsize=8)

# Add an annotation for the slope
if slope != float('inf'):
    slope_annotation = f"Slope: {slope:.2f}"
else:
    slope_annotation = "Slope: Undefined"
ax.annotate(slope_annotation, ((x1 + x2) / 2, (y1 + y2) / 2), textcoords="offset points", xytext=(-15, 10), ha='center', fontsize=8)


# Add an annotation for the slope
slope_annotation = f"Slope: {slope:.2f}"
ax.annotate(slope_annotation, ((x1 + x2) / 2, (y1 + y2) / 2), textcoords="offset points", xytext=(-15, 10), ha='center', fontsize=8)

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Line Plot')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```
